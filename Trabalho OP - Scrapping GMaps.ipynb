```python
# Bibliotecas

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import win32com.client #Executar no prompt Anaconda: pip install pywin32
```

```python
# Carregando arquivo com a lista de Endereços e de Castelos

Enderecos = pd.read_excel('Trabalho OP.xlsx', sheet_name = 'Endereços', nrows = 25)
Castelos = pd.read_excel('Trabalho OP.xlsx', sheet_name = 'Dados de Entrada', nrows = 25)
```

```python
# Instancia Excel
File = win32com.client.Dispatch("Excel.Application")

File.Visible = 1

# Abre o arquivo
Workbook = File.Workbooks.open("C:\\Users\\Thais\\Documents\\BI Master\\OP - Trabalho\\Trabalho OP.xlsx")
```

```python
# Inicializando o Chrome com definição dos parâmetros

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--start-maximized')
wd = webdriver.Chrome(r"chromedriver.exe",options=chrome_options)
```

```python
# Limpando cookies

wd.delete_all_cookies()
```

```python
# Abrindo Maps

wd.get("https://www.google.com.br/maps")
time.sleep(4)
```

```python
# Scrapping

linha = Castelos.columns[1:]
coluna = Castelos['From\To']

# Clica no botão de rota
botaoRota = wd.find_element_by_xpath('//*[@id="hArJGc"]/div')
botaoRota.click()
time.sleep(2)

for c in range(len(linha)):
  casteloDestino = Enderecos['Endereço'][c]
  
  for i in range(0,2):
    try:
      # Preenche campo de pesquisa (endereço destino)
      enderecoDestino = wd.find_element_by_xpath('//*[@id="sb_ifc51"]/input')
      enderecoDestino.clear()
      enderecoDestino.send_keys(casteloDestino)
      enderecoDestino.send_keys(Keys.ENTER)
      time.sleep(3)
      erro = None
    except Exception as e:
      erro = str(e)

    if erro:
      time.sleep(2)
    else:
      break  
  
  for l in range(len(coluna)):
    if l == c:
      # Grava no excel na célula correspondente
      File.Sheets(2).Cells(l+2, c+2).Value = float(0)
    else:
      casteloOrigem = Enderecos['Endereço'][l]
      
      for i in range(0,5):
        try:
          # Preenche endereço de origem
          enderecoOrigem = wd.find_element_by_xpath('//*[@id="sb_ifc50"]/input')
          enderecoOrigem.clear()
          enderecoOrigem.send_keys(casteloOrigem)
          enderecoOrigem.send_keys(Keys.ENTER)
          time.sleep(2)

          # Clica no botão "Carro"
          botaoCarro = wd.find_element_by_xpath('//*[@id="omnibox-directions"]/div/div[2]/div/div/div/div[2]/button')
          botaoCarro.click()
          time.sleep(2)

          # Recupera distância entre os castelos
          distancia = wd.find_element_by_xpath('//*[@id="section-directions-trip-0"]/div[1]/div[1]/div[1]/div[2]/div').text[:-7]

          # Grava no excel na célula correspondente
          File.Sheets(2).Cells(l+2, c+2).Value = float(distancia.replace(',', '.'))
          # Range("Início:Fim")
          # Workbook.RefreshAll()
          time.sleep(1)
          erro = None
        except Exception as e:
          erro = str(e)
          if i == 4:
            print("Distância não encontrada para o trecho: ", linha[c], " -> ", coluna[l])

        if erro:
          time.sleep(2)
        else:
          break
```

```python
# Salve e fecha o excel
Workbook.Save()
File.Quit()

# Encerra o Scrapping
wd.quit()
```
